```python
from datasets import load_dataset
from transformers import BertTokenizer
import pandas as pd

checkpoint = "neuralmind/bert-base-portuguese-cased"
num_proc = 5

pd.set_option("display.max_colwidth", 10000)

```

```python
tokenizer = BertTokenizer.from_pretrained(checkpoint)
legal_text = load_dataset("pierreguillou/lener_br_finetuning_language_model")
legal_text = legal_text.map(lambda x: tokenizer(x["text"], add_special_tokens=False),num_proc=num_proc,batched=True)

wikipedia = load_dataset("jvanz/portuguese_wikipedia_sentences")
wikipedia = wikipedia.map(lambda x: tokenizer(x["text"], add_special_tokens=False),num_proc=num_proc,batched=True)
```
